```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.cluster_centers_
```

## Evaluation


```python
def privOrNot(x):
    if(x=='Yes'):
        return 1
    else:
        return 0
df['Cluster']=df['Private'].apply(privOrNot)
```

```python
kmeans.labels_
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(df['Cluster'], kmeans.labels_))
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores=[]
for i in range(2,11):
    kmeans=KMeans(n_clusters=i, n_init=10)
    cluster_labels=kmeans.fit_predict(df.drop('Private', axis=1)) # .fit_predict returns labels in array format, .fit returns trained model as object
    
    silhouette_avg=silhouette_score(df.drop('Private', axis=1), cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
plt.plot(range(2,11), silhouette_scores, marker='o', color='blue', mfc='red')

    
    
```
